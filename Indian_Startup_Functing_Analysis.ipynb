```python
# Data Visulization using Matplotlib Library
# 1) How has startup funding trend changed over the years?
import matplotlib.pyplot as plt
demo = data.groupby("year")["amount"].sum()
plt.plot(demo.index, demo.values, marker='o', markerfacecolor="red",color="black",markeredgecolor="green",linestyle="--",linewidth=1.5,label="Total Amount")
plt.xlabel("Year",color="blue",fontsize=12)
plt.ylabel("Amount",color="Olive",fontsize=12)
plt.title("Startup Funding Trend Over the Years",color="red",fontsize=15)
plt.grid(linestyle=":", color="skyblue")

plt.legend()
plt.show()
```

```python
# 2)which city has highest number of startup
import matplotlib.pyplot as plt


demo = data["city"].value_counts().head(10)
color = ["#FF84BA","#FFDF82","#99C2FF","#FFEFE3","#FF9E20","#BD4444","#73976A","#BBDC12","#FF80C7","#17433F"]
plt.bar(demo.index,demo.values,color = color,edgecolor = 'black')

plt.title("Top 10 startup companies in India",color='olive',fontsize=16)
plt.xticks(rotation=60)
plt.xlabel("City",color='Blue',fontsize=12)
plt.ylabel("Number of Startups",color='Blue',fontsize=12)
# plt.grid(color="Green",linestyle="--",linewidth=1.5)
plt.show()
```

```python
#3) Which startup sector have the higher number of funded startups?
import matplotlib.pyplot as plt

demo = data['vertical'].value_counts().head(10)
color = ["#FF84BA","#FFDF82","#99C2FF","#FFEFE3","#FF9E20","#BD4444","#73976A","#BBDC12","#FF80C7","#17433F"]

plt.figure(figsize=(8,5))

plt.barh(demo.index,demo.values[::-1],color = color,edgecolor = 'black')
plt.xticks(rotation=60)
plt.title("Top 10 funded startups",color='olive',fontsize=16)
plt.xlabel("Verticle",color='#BD4444',fontsize=15)
plt.ylabel("Number of Startups",color='#BD4444',fontsize=15)

plt.show()
```

```python
# 4) What is the distribution of different funding rounds?
import matplotlib.pyplot as plt

demo = data['round'].value_counts().head(7)
color = ["#FF84BA","#FFDF82","#99C2FF","#FFEFE3","#FF9E20","#BD4444","#73976A"]

plt.figure(figsize=(10,6))
plt.figure(figsize=(10,6))

plt.pie(demo.values,labels=demo.index,autopct='%1.1f%%')
# plt.xticks(rotation=60)
plt.title("Funding rounds distributions in percentage",color='#FF80C7',fontsize=16)
# plt.xlabel("Vertical",color='#BD4444',fontsize=15)
# plt.ylabel("Number of Rounds",color='#BD4444',fontsize=15)
```

```python
#5) How are startup funding amounts distribution?
import matplotlib.pyplot as plt

demo=data['amount'].value_counts().head(10)
demo
plt.hist(demo.values, bins=10, edgecolor = "black")
plt.title("Distribution Of Startup Funding")
```

```python
#6) The relationship between Funding year and funcding Amount?
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(data["year"],data["amount"])
plt.title("Funding Year and Funding Amount")
plt.xlabel("Funding Year")
plt.ylabel("Funding Amount")
```

```python
# 7) Which funding round occurs most frequently?
# import matplotlib.pyplot as plt
import seaborn as sns

demo = data['round'].value_counts().head(5).index
sns.countplot(data, y='round',order= demo,palette="viridis")
plt.xticks(rotation=60)
plt.title("Funding Round Distribution")
plt.xlabel("Funding Year")
plt.ylabel("Funding Amount")
```
